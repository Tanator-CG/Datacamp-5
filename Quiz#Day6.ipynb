```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#QUiz_1
name = ['A','B','C','D']
year = ['2014','2015','2016','2017']
revenue = np.random.random_sample((4,4))+np.random.randint(50,501,(4,4))
df = pd.DataFrame(data=revenue,index=name,columns=year)
df
```

```python
#Quiz_2
fig = plt.figure(figsize=(5,5))
fig = sns.barplot(x=df.index,y=df.mean().values)
fig2 = plt.figure(figsize=(5,5))
fig2 = sns.barplot(x=df.mean().values,y=df.index)
```

```python
#Quiz_3
fig = plt.figure(figsize=(5,5))
fig = sns.scatterplot(x=df.index, y='2017',data=df)
fig2 = plt.figure(figsize=(5,5))
fig2 = sns.scatterplot(x='2017', y=df.index,data=df)
```

```python
#Quiz_13
fig = plt.figure(figsize=(5,5))
new_df = pd.crosstab(df['Pclass'],df['Sex'])
sns.heatmap(new_df,cmap='Greens',linewidth=1.5)
```
